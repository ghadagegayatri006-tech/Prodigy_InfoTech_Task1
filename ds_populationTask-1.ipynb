## Dropping unnecessary columns

```python
df.drop(['Indicator Name','Country Name'],axis=1,inplace=True)
```

## Extraction of top 10 countries with respect to total population

```python
#Filtering data for total population
total_population = df[df['Indicator Code'] == 'SP.POP.TOTL']

#Sort data based on the total population for 2024
total_population_sorted=total_population.sort_values(by='2024',ascending=False)

#Get top 10 countries with the highest population for 2024
total_top10_countries=total_population_sorted.head(10)


print("total_top10_countries of total population \n")
print(total_top10_countries[['Country Code']])
```

## Histogram (continuous variable)

## Population of Selected Countries in 2024

```python
# List of countries for the graph
countries_to_plot = ['USA', 'CAN', 'BRA', 'JPN', 'DEU', 'FRA']

# Convert 2024 column to numeric
df['2024'] = pd.to_numeric(df['2024'], errors='coerce')

# Filter the DataFrame for selected countries
filtered_data = df[df['Country Code'].isin(countries_to_plot)]

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(filtered_data['Country Code'], filtered_data['2024'])
plt.xlabel('Country Code')
plt.ylabel('Population in 2024')
plt.title('Population of Selected Countries in 2024')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Bar Plot

## Top ten countries of total population in year 2024 and 2018

```python
#Create a bar plot
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
sns.barplot(x='2024',y='Country Code',data=total_top10_countries, palette='Blues_r')
plt.title('Top 10 Countries by Total Population in (2024)', fontsize=10)
plt.xlabel('Total population', fontsize=10)

plt.ylabel('Country', fontsize=10)
plt.show()



#Create a bar plot
plt.figure(figsize=(12,6))
plt.subplot(2,2,2)
sns.barplot(x='2018',y='Country Code',data=total_top10_countries, palette='Blues_r')
plt.title('Top 10 Countries by Total Population in (2018)', fontsize=10)
plt.xlabel('Total population', fontsize=10)
plt.ylabel('Country', fontsize=10)
plt.show()
```
